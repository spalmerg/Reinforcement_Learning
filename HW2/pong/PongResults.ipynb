```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Final Version
- 2 episodes per epoch
- 0.0001 Learning Rate
- Hidden Layer Size = 200

```python
filepath = open('PongTraining3.txt')
text = filepath.read().splitlines()
text = text[10:]
rewards = []
for i in range(len(text)):
    rewards.append(float(text[i].split()[4]))
    
plt.plot(rewards)
plt.title('Pong--48 Hours')
```
